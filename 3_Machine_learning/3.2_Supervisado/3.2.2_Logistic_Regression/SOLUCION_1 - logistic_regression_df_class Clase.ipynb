# Ejercicio Python de Regresión Logística
Realizaremos un ejercicio de prueba para comprender como funciona este algoritmo

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Cargamos los datos de entrada del archivo csv

```python
dataframe = pd.read_csv("Ficheros/usuarios_win_mac_lin.csv")
dataframe.head()
```

```python
dataframe.paginas.value_counts()
```

```python
dataframe.describe()
```

```python
print(dataframe.groupby('clase').size())
```

## Visualizamos los datos

```python
dataframe.drop(labels=['clase'], axis=1).hist() # 1 es la columna, y no la estamos quitando para siempre
plt.tight_layout()
```

```python
sns.pairplot(dataframe.dropna(),
            hue='clase',
            height=4,
            vars=["duracion", "paginas", "acciones", "valor"],
            kind='reg'); # reg es regresión lineal
```

## Creamos el modelo

```python
dataframe.drop(labels=['clase'], axis=1)
```

```python
# en este caso no dividimos en train_test. Es un ejemplo.
X = np.array(dataframe.drop(labels=['clase'], axis=1))
y = np.array(dataframe['clase'])
X.shape
```

```python
from sklearn import linear_model

model = linear_model.LogisticRegression(solver='liblinear', multi_class = 'auto')
model.fit(X,y)
```

```python
predictions = model.predict(X)
predictions
```

```python
predict_probab = model.predict_proba(X)
np.round(predict_probab, 2)
```

```python
model.score(X, y) # accuracy
```

# Adicional: Validación del Modelo

```python
validation_size = 0.2
seed = 42
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,
                                                                   y,
                                                                    test_size=validation_size,
                                                                   random_state = seed)
```

```python
name = 'Logistic Regression'
kfold = model_selection.KFold(n_splits=10) # 10 folds para validación cruzada
cv_results = model_selection.cross_val_score(model, X_train, y_train, cv=kfold, scoring='accuracy')

msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(cv_results)
print(msg)
```

```python
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print(accuracy_score(y_test, predictions))
```

```python
model.predict(X_test)
```

```python
acierto = accuracy_score(y_test, predictions)
error = 1 - acierto
print("Acierto:", round(acierto*100,2), "%")
print("Error:", round(error*100,2), "%")
```

## Reporte de Resultados

```python
c_matrix = confusion_matrix(y_test, predictions)
c_matrix
```

```python
import seaborn as sns
sns.heatmap(c_matrix, annot=True)
```

```python
sns.heatmap(c_matrix/c_matrix.sum(), annot=True, cmap='Blues', fmt = '.2%')
```
